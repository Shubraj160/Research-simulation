```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import random
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Flatten, Dense, Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import SGD
from concurrent.futures import ThreadPoolExecutor
```

```python
# 1. Thread Configuration for Concurrency
tf.config.threading.set_intra_op_parallelism_threads(1)
tf.config.threading.set_inter_op_parallelism_threads(1)
print("Environment set. TF restricted to 1 thread per instance to allow concurrency.")
```

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
X_train = X_train.astype(np.float32) / 255.0
X_test = X_test.astype(np.float32) / 255.0
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)
```

```python
def iid_partition(x, y, num_clients):
    idx = np.random.permutation(len(x))
    x_shuffled, y_shuffled = x[idx], y[idx]
    x_parts = np.array_split(x_shuffled, num_clients)
    y_parts = np.array_split(y_shuffled, num_clients)
    partitions = list(zip(x_parts, y_parts))
    return partitions
```

```python
def non_iid_partition(x, y, num_clients):
    sorted_idx = np.argsort(y)
    X_sorted, Y_sorted = x[sorted_idx], y[sorted_idx]
    num_shards = 2 * num_clients
    size_shards = len(x) // num_shards
    shards_idx = np.random.permutation(num_shards)
    partitions = []
    for i in range(num_clients):
        s1, s2 = shards_idx[2*i], shards_idx[2*i+1]
        X_client = np.concatenate((X_sorted[s1*size_shards:(s1+1)*size_shards], X_sorted[s2*size_shards:(s2+1)*size_shards]), axis=0)
        Y_client = np.concatenate((Y_sorted[s1*size_shards:(s1+1)*size_shards], Y_sorted[s2*size_shards:(s2+1)*size_shards]), axis=0)
        idx = np.random.permutation(len(X_client))
        partitions.append((X_client[idx], Y_client[idx]))
    return partitions
```

```python
# 4. Model Generation
def cnn_model():
    model = Sequential([
        Input(shape=(28, 28, 1)),
        Conv2D(32, kernel_size=(3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(64, kernel_size=(3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax'),
    ])
    model.compile(optimizer=SGD(learning_rate=0.01), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
def local_training(client_data, global_weights, mu):
    X_local, Y_local = client_data
    num_sample = len(X_local)
    model = cnn_model()
    model.set_weights(global_weights)
    optimizer = SGD(learning_rate=0.01)
    loss_fn = tf.keras.losses.SparseCategoricalCrossentropy()
    global_weights_tensors = [tf.constant(w) for w in global_weights]
    dataset = tf.data.Dataset.from_tensor_slices((X_local, Y_local)).batch(32)
    
    def train_step(x, y):
        with tf.GradientTape() as tape:
            y_pred = model(x, training=True)
            base_loss = loss_fn(y, y_pred)
            
            proximal_term = 0.0
            for local_var, global_var in zip(model.trainable_variables, global_weights_tensors):
                proximal_term += tf.reduce_sum(tf.square(local_var - global_var))
                
            total_loss = base_loss + (mu / 2.0) * proximal_term    
        grads = tape.gradient(total_loss, model.trainable_variables)
        optimizer.apply_gradients(zip(grads, model.trainable_variables))

    for epoch in range(2):
        for batch_x, batch_y in dataset:
            train_step(batch_x, batch_y)
            
    return model.get_weights(), num_sample
               
```

```python
def aggregate_weights(client_updates, global_weights):
    tot_sample = sum(num_samples for _, num_samples in client_updates)
    new_weights = [np.zeros_like(w) for w in global_weights]
    
    for local_weights, num_samples in client_updates:
        w_factor = num_samples / tot_sample
        for i in range(len(new_weights)):
            new_weights[i] += local_weights[i] * w_factor
            
    return new_weights
```

```python
def evaluate_global(global_model, X_test, Y_test):
    loss, accuracy = global_model.evaluate(X_test, Y_test, verbose=1)
    return loss, accuracy
```

```python
def run_simulation(partitions, X_test, Y_test, num_clients, rounds, participation_rate, mu):
    global_model = cnn_model()
    global_weights = global_model.get_weights()
    max_workers = min(num_clients, 12)
    history = []
    
    for r in range(1, rounds + 1):
        num_active = max(1, int(num_clients * participation_rate))
        active_clients = random.sample(range(num_clients), num_active)
        client_updates = []
        
        with ThreadPoolExecutor(max_workers=max_workers) as executor:
            futures = [executor.submit(local_training, partitions[i], global_weights, mu) for i in active_clients]
            for future in futures:
                client_updates.append(future.result())
                
        # Aggregate and Update
        global_weights = aggregate_weights(client_updates, global_weights)
        global_model.set_weights(global_weights)
        
        # Evaluate
        loss, acc = evaluate_global(global_model, X_test, Y_test)
        history.append({'round': r, 'loss': loss, 'acc': acc})
        print(f"Round {r} | Accuracy: {acc} | Loss: {loss}")
        
    return history
```

```python
CLIENTS=10
ROUNDS_IID=20
ROUNDS_NON_IID=50
PARTICIPATION_RATE=0.5
MU1=0.1
MU2=0.5
    
print(f"=== RUNNING IID FEDPROX ({MU1} Proximal Coefficient, {PARTICIPATION_RATE*100}% Participation) ===")
iid_parts = iid_partition(X_train, Y_train, CLIENTS)
iid_history_mu1 = run_simulation(iid_parts, X_test, Y_test, CLIENTS, ROUNDS_IID, PARTICIPATION_RATE, MU1)

print(f"=== RUNNING IID FEDPROX ({MU2} Proximal Coefficient, {PARTICIPATION_RATE*100}% Participation) ===")
iid_history_mu2 = run_simulation(iid_parts, X_test, Y_test, CLIENTS, ROUNDS_IID, PARTICIPATION_RATE, MU2)
    
print(f"\n=== RUNNING NON-IID FEDPROX ({MU1} Proximal Coefficient, {PARTICIPATION_RATE*100}% Participation) ===")
non_iid_parts = non_iid_partition(X_train, Y_train, CLIENTS)
non_iid_history_mu1 = run_simulation(non_iid_parts, X_test, Y_test, CLIENTS, ROUNDS_NON_IID, PARTICIPATION_RATE, MU1)

print(f"\n=== RUNNING NON-IID FEDPROX ({MU2} Proximal Coefficient, {PARTICIPATION_RATE*100}% Participation) ===")
non_iid_history_mu2 = run_simulation(non_iid_parts, X_test, Y_test, CLIENTS, ROUNDS_NON_IID, PARTICIPATION_RATE, MU2)
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# 1. Convert all four histories into Pandas DataFrames
iid_df_mu1 = pd.DataFrame(iid_history_mu1)
iid_df_mu2 = pd.DataFrame(iid_history_mu2)

non_iid_df_mu1 = pd.DataFrame(non_iid_history_mu1)
non_iid_df_mu2 = pd.DataFrame(non_iid_history_mu2)

# ==========================================
# GRAPH 1: IID Data (MU1 vs MU2)
# ==========================================
plt.figure(figsize=(10, 5))

# MU1 Lines
plt.plot(iid_df_mu1['round'], iid_df_mu1['acc'], marker='o', color='blue', label=f'Test Acc (μ={MU1})')
plt.plot(iid_df_mu1['round'], iid_df_mu1['loss'], marker='s', color='lightblue', linestyle='--', label=f'Test Loss (μ={MU1})')

# MU2 Lines
plt.plot(iid_df_mu2['round'], iid_df_mu2['acc'], marker='^', color='green', label=f'Test Acc (μ={MU2})')
plt.plot(iid_df_mu2['round'], iid_df_mu2['loss'], marker='d', color='lightgreen', linestyle='--', label=f'Test Loss (μ={MU2})')

plt.title(f"FedProx (IID) - {CLIENTS} Clients")
plt.xlabel("Communication Rounds")
plt.ylabel("Value")
# Using ROUNDS_IID (20 rounds). Step by 2.
plt.xticks(range(0, ROUNDS_IID + 1, 2))
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()

# ==========================================
# GRAPH 2: Non-IID Data (MU1)
# ==========================================
plt.figure(figsize=(10, 5))

# MU1 Lines
plt.plot(non_iid_df_mu1['round'], non_iid_df_mu1['acc'], marker='o', color='blue', label=f'Test Acc (μ={MU1})')
plt.plot(non_iid_df_mu1['round'], non_iid_df_mu1['loss'], marker='s', color='lightblue', linestyle='--', label=f'Test Loss (μ={MU1})')

plt.title(f"FedProx (Non-IID) - μ={MU1} - {CLIENTS} Clients")
plt.xlabel("Communication Rounds")
plt.ylabel("Value")
# Using ROUNDS_NON_IID (50 rounds). Step by 5 to avoid crowding the x-axis.
plt.xticks(range(0, ROUNDS_NON_IID + 1, 5))
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()

# ==========================================
# GRAPH 3: Non-IID Data (MU2)
# ==========================================
plt.figure(figsize=(10, 5))

# MU2 Lines (Changed to green to match the IID graph legend)
plt.plot(non_iid_df_mu2['round'], non_iid_df_mu2['acc'], marker='^', color='green', label=f'Test Acc (μ={MU2})')
plt.plot(non_iid_df_mu2['round'], non_iid_df_mu2['loss'], marker='d', color='lightgreen', linestyle='--', label=f'Test Loss (μ={MU2})')

plt.title(f"FedProx (Non-IID) - μ={MU2} - {CLIENTS} Clients")
plt.xlabel("Communication Rounds")
plt.ylabel("Value")
# Using ROUNDS_NON_IID (50 rounds). Step by 5 to avoid crowding the x-axis.
plt.xticks(range(0, ROUNDS_NON_IID + 1, 5))
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()
```
