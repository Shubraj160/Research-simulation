```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import random
from tensorflow.keras.datasets import fashion_mnist  # Changed from mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Flatten, Dense, Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import SGD
from concurrent.futures import ThreadPoolExecutor
```

```python
tf.config.threading.set_intra_op_parallelism_threads(1)
tf.config.threading.set_inter_op_parallelism_threads(1)
print("Environment set. TF restricted to 1 thread per instance to allow 12-thread concurrency.")
```

```python
(X_train, Y_train), (X_test, Y_test) = fashion_mnist.load_data()
```

```python
X_train = X_train.astype(np.float32) / 255.0
X_test = X_test.astype(np.float32) / 255.0
```

```python
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)
```

```python
def iid_partition(x, y, num_clients):
    idx = np.random.permutation(len(x))
    x_shuffled, y_shuffled = x[idx], y[idx]
    x_parts = np.array_split(x_shuffled, num_clients)
    y_parts = np.array_split(y_shuffled, num_clients)
    partitions = list(zip(x_parts, y_parts))
    return partitions
```

```python
def non_iid_partition(x, y, num_clients):
    sorted_idx = np.argsort(y)
    X_sorted, Y_sorted = x[sorted_idx], y[sorted_idx]
    num_shards = 2 * num_clients
    size_shards = len(x) // num_shards
    shards_idx = np.random.permutation(num_shards)
    partitions = []
    for i in range(num_clients):
        s1, s2 = shards_idx[2*i], shards_idx[2*i+1]
        X_client = np.concatenate((X_sorted[s1*size_shards:(s1+1)*size_shards], X_sorted[s2*size_shards:(s2+1)*size_shards]), axis=0)
        Y_client = np.concatenate((Y_sorted[s1*size_shards:(s1+1)*size_shards], Y_sorted[s2*size_shards:(s2+1)*size_shards]), axis=0)
        idx = np.random.permutation(len(X_client))
        partitions.append((X_client[idx], Y_client[idx]))
    return partitions
```

```python
def cnn_model():
    model = Sequential([
        Input(shape=(28, 28, 1)),
        Conv2D(32, kernel_size=(3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(64, kernel_size=(3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax'), # Remains 10 classes (T-shirts, trousers, bags, etc.)
    ])
    model.compile(optimizer=SGD(learning_rate=0.01), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
def local_training(client_data, global_weights):
    X_local, Y_local = client_data
    num_sample = len(X_local)
    model = cnn_model()
    model.set_weights(global_weights)
    model.fit(X_local, Y_local, epochs=2, batch_size=32, verbose=0)
    local_weights = model.get_weights()
    return local_weights, num_sample
```

```python
class federated_server:
    def __init__(self, X_test, Y_test):
        self.global_model = cnn_model()
        self.X_test = X_test
        self.Y_test = Y_test
    def get_global(self):
        return self.global_model.get_weights()
    def set_global(self, weights):
        self.global_model.set_weights(weights)
    def aggregate_weights(self, client_updates):
        tot_sample = sum(num_samples for _, num_samples in client_updates)
        global_weights = self.get_global()
        new_weights = [np.zeros_like(w) for w in global_weights]
        for local_weights, num_samples in client_updates:
            w_factor = num_samples / tot_sample
            for i in range(len(new_weights)):
                new_weights[i] += local_weights[i] * w_factor
        self.set_global(new_weights)
    def evaluate_global(self):
        loss, accuracy = self.global_model.evaluate(self.X_test, self.Y_test, verbose=1)
        return loss, accuracy

def run_simulation(partitions, X_test, Y_test, num_clients, rounds, participation_rate):
    server = federated_server(X_test, Y_test)
    max_workers = min(num_clients, 12)
    history = []
    for r in range(1, rounds + 1):
        global_weights = server.global_model.get_weights()
        num_active = max(1, int(num_clients * participation_rate))
        active_clients = random.sample(range(num_clients), num_active)
        client_updates = []
        with ThreadPoolExecutor(max_workers=max_workers) as executor:
            futures = [executor.submit(local_training, partitions[i], global_weights) for i in active_clients]
            for future in futures:
                client_updates.append(future.result())
        server.aggregate_weights(client_updates)
        loss, acc = server.evaluate_global()
        history.append({
            'round': r,
            'loss': loss,
            'acc': acc
        })
        print(f"Round {r} | Accuracy: {acc} | Loss: {loss}")
    return history
```

```python
CLIENTS=10
IID_ROUNDS=50
PARTICIPATION_RATE=1
print(f"===RUNNING IID FEDERATED LEARNING 100% PARTICIPATION({CLIENTS} Clients)===")
iid_parts=iid_partition(X_train,Y_train,CLIENTS)
iid_history1=run_simulation(iid_parts,X_test,Y_test,CLIENTS,IID_ROUNDS,PARTICIPATION_RATE)
```

```python
NON_IID_ROUNDS=50
non_iid_parts=non_iid_partition(X_train,Y_train,CLIENTS)
print(f"===RUNNING NON-IID FEDERATED LEARNING 100% PARTICIPATION({CLIENTS} Clients)===")
non_iid_history1=run_simulation(non_iid_parts,X_test,Y_test,CLIENTS,NON_IID_ROUNDS,PARTICIPATION_RATE)
```

```python
IID_ROUNDS=50
PARTICIPATION_RATE=0.5
print(f"===RUNNING IID FEDERATED LEARNING 50% PARTICIPATION({CLIENTS} Clients)===")
iid_history2=run_simulation(iid_parts,X_test,Y_test,CLIENTS,IID_ROUNDS,PARTICIPATION_RATE)
NON_IID_ROUNDS=100
print(f"===RUNNING NON-IID FEDERATED LEARNING 50% PARTICIPATION({CLIENTS} Clients)===")
non_iid_history2=run_simulation(non_iid_parts,X_test,Y_test,CLIENTS,NON_IID_ROUNDS,PARTICIPATION_RATE)
```

```python
IID_ROUNDS=50
PARTICIPATION_RATE=0.3
print(f"===RUNNING IID FEDERATED LEARNING 30% PARTICIPATION({CLIENTS} Clients)===")
iid_history3=run_simulation(iid_parts,X_test,Y_test,CLIENTS,IID_ROUNDS,PARTICIPATION_RATE)
NON_IID_ROUNDS=100
print(f"===RUNNING NON-IID FEDERATED LEARNING 30% PARTICIPATION({CLIENTS} Clients)===")
non_iid_history3=run_simulation(non_iid_parts,X_test,Y_test,CLIENTS,NON_IID_ROUNDS,PARTICIPATION_RATE)
```

```python
np.save("F_mnist_iid_history1.npy", iid_history1)
np.save("F_mnist_non_iid_history1.npy", non_iid_history1)

np.save("F_mnist_iid_history2.npy", iid_history2)
np.save("F_mnist_non_iid_history2.npy", non_iid_history2)

np.save("F_mnist_iid_history3.npy", iid_history3)
np.save("F_mnist_non_iid_history3.npy", non_iid_history3)
```

```python

def style_plot(ax, max_rounds, step):
    ax.set_xticks(range(0, max_rounds + 1, step))
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color('#CCCCCC')
    ax.spines['bottom'].set_color('#CCCCCC')
    
   
    ax.grid(axis='y', color='#E0E0E0', linestyle='-')
    ax.grid(axis='x', visible=False)
    
   
    ax.tick_params(axis='both', which='both', length=0)
    
   
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.18), ncol=2, frameon=False)


# ==========================================
# 100% PARTICIPATION PLOTS
# ==========================================

# --- 100% IID Plot (50 Rounds, Spacing: 2) ---
fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(iid_df1['round'], iid_df1['acc'], color='#4472C4', label='Test Accuracy', linewidth=2)
ax.plot(iid_df1['round'], iid_df1['loss'], color='#ED7D31', label='Test Loss', linewidth=2)
ax.set_title("FedAvg Global Performance-IID (100% Participation)", pad=15, fontsize=12)
ax.set_xlabel("Communication Rounds")
ax.set_ylabel("Value")
style_plot(ax, max_rounds=50, step=5)
plt.tight_layout()
plt.show()

# --- 100% Non-IID Plot (50 Rounds, Spacing: 5) ---
fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(non_iid_df1['round'], non_iid_df1['acc'], color='#4472C4', label='Test Accuracy', linewidth=2)
ax.plot(non_iid_df1['round'], non_iid_df1['loss'], color='#ED7D31', label='Test Loss', linewidth=2)
ax.set_title("FedAvg Global Performance-Non-IID (100% Participation)", pad=15, fontsize=12)
ax.set_xlabel("Communication Rounds")
ax.set_ylabel("Value")
style_plot(ax, max_rounds=50, step=5)
plt.tight_layout()
plt.show()


# ==========================================
# 50% PARTICIPATION PLOTS
# ==========================================

# --- 50% IID Plot (50 Rounds, Spacing: 5) ---
fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(iid_df2['round'], iid_df2['acc'], color='#4472C4', label='Test Accuracy', linewidth=2)
ax.plot(iid_df2['round'], iid_df2['loss'], color='#ED7D31', label='Test Loss', linewidth=2)
ax.set_title("FedAvg Global Performance-IID (50% Participation)", pad=15, fontsize=12)
ax.set_xlabel("Communication Rounds")
ax.set_ylabel("Value")
style_plot(ax, max_rounds=50, step=5)
plt.tight_layout()
plt.show()

# --- 50% Non-IID Plot (100 Rounds, Spacing: 10) ---
fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(non_iid_df2['round'], non_iid_df2['acc'], color='#4472C4', label='Test Accuracy', linewidth=2)
ax.plot(non_iid_df2['round'], non_iid_df2['loss'], color='#ED7D31', label='Test Loss', linewidth=2)
ax.set_title("FedAvg Global Performance-Non-IID (50% Participation)", pad=15, fontsize=12)
ax.set_xlabel("Communication Rounds")
ax.set_ylabel("Value")
style_plot(ax, max_rounds=100, step=10)
plt.tight_layout()
plt.show()


# ==========================================
# 30% PARTICIPATION PLOTS
# ==========================================

# --- 30% IID Plot (50 Rounds, Spacing: 5) ---
fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(iid_df3['round'], iid_df3['acc'], color='#4472C4', label='Test Accuracy', linewidth=2)
ax.plot(iid_df3['round'], iid_df3['loss'], color='#ED7D31', label='Test Loss', linewidth=2)
ax.set_title("FedAvg Global Performance-IID (30% Participation)", pad=15, fontsize=12)
ax.set_xlabel("Communication Rounds")
ax.set_ylabel("Value")
style_plot(ax, max_rounds=50, step=5)
plt.tight_layout()
plt.show()

# --- 30% Non-IID Plot (100 Rounds, Spacing: 10) ---
fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(non_iid_df3['round'], non_iid_df3['acc'], color='#4472C4', label='Test Accuracy', linewidth=2)
ax.plot(non_iid_df3['round'], non_iid_df3['loss'], color='#ED7D31', label='Test Loss', linewidth=2)
ax.set_title("FedAvg Global Performance-Non-IID (30% Participation)", pad=15, fontsize=12)
ax.set_xlabel("Communication Rounds")
ax.set_ylabel("Value")
style_plot(ax, max_rounds=100, step=10)
plt.tight_layout()
plt.show()
```
